popularity based recommender system

```python
#top 50 books with highest average rating
#consider only those books which have minumum 200 votes 
```

```python
#merge book with rating
#both are merged on basis of isbn column
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
#first,we group the dataset by "book-title" column then,count occurences of book rating
#reset_index will get the dataframe
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df
```

```python
#rename column name book-rating to num_ratings(number of ratings)
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
#book-title with number of ratings on each book
```

```python
#average rating of each book
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
#merging both to get number of ratings and average rating
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
#those books which have more than 200 number of ratings 
popular_df = popular_df[popular_df['num_ratings']>=200]
```

```python
popular_df
```

```python
#getting top 50 books which have highest average rating
popular_df=popular_df.sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
#we will use the merged dataset of books and ratings
ratings_with_name.head()
```

```python
#filtering of those users who have rated atleast 200 books
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200 
```

```python
#boolean indexing of those users whose have rated atleast 200 boooks and finding their index
experienced_users=x[x].index
#there are 811 experienced users
```

```python
#fetchings only those those from ratings_with_name in which experienced user is there
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(experienced_users)]
```

```python
#only those books wich have atleast 50 number of ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
y
```

```python
famous_books = y[y].index
```

```python
famous_books
#there are 706 famous books
```

```python
#fetching only those values form filtered rating where famous book is there
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#drop dupliactes if there
final_ratings=final_ratings.drop_duplicates()
```

```python
final_ratings
#removing duplicate book-tile entries as in next step,a grid is formed.grid cannot form of duplicated entries.
final_ratings.drop_duplicates(subset='Book-Title', inplace=True)
```

```python
#now making a n*n grid where rows are book-title,columns are user-id and values are ratings
pt=final_ratings.pivot(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt
```

```python
#replacing nan with 0
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#above points are vectors and we will find similar vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]#index of a particular book
    #finding most similar books x:x[1]-on the basis of similarity score
    #emurate:book number and thier similarity score
    #[1:5] will return the top similar book
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]#getting name
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```

```python
recommend('Message in a Bottle')
```

```python
recommend('The Notebook')
```

```python
recommend('Animal Farm')
```
